# Handling Outliers

outlier data replace with upperbound or lowerbound

```python
def outliers(df, column):
  q1 = df[column].quantile(0.25)
  q3 = df[column].quantile(0.75)
  iqr = q3-q1
  lower_bound = q1-1.5*iqr
  upper_bound = q3+1.5*iqr

  for i in range(len(df)):
      if df[column].iloc[i] > upper_bound:
          df[column].iloc[i] = upper_bound
      if df[column].iloc[i] < lower_bound:
          df[column].iloc[i] = lower_bound
```

```python
columns = list(df.columns)
columns = columns[1:]

for column in columns:
  outliers(df, column)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data_train = df[columns]
data_train[columns] = scaler.fit_transform(data_train)
```

```python
data_train.head()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=100).fit(data_train)
labels = kmeans.labels_
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Ambil fitur numerik dan kategorikal
numerical_features = data_train.select_dtypes(include=['float64', 'int64'])
categorical_features = data_train.select_dtypes(include=['object'])

# Lakukan one-hot encoding pada fitur-fitur kategorikal
encoded_categorical_features = pd.get_dummies(categorical_features)

# Gabungkan kembali fitur-fitur numerik dan hasil one-hot encoding
encoded_data_train = pd.concat([numerical_features, encoded_categorical_features], axis=1)

wcss=[]
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=40)
    kmeans.fit(encoded_data_train)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(12,8))
plt.plot(range(1, 11), wcss)
plt.scatter(range(1, 11), wcss)

plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
